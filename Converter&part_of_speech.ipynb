```python
file_name1 = "testvideo.mp4"  # @param {type:"string"}
file_type = "video"  # @param ["audio","video"]
model_size = "small"  # @param ["base", "small", "medium", "large"]

# @markdown <font size="4">Run this block, and your file will be turned into subtitle with timestamp directly.

import whisper
import time
import ffmpeg
import os
import subprocess



model = whisper.load_model(model_size)
result = model.transcribe(file_name1)
#print(result["text"])


for item in result["segments"]:
    minutes = int(list(item.values())[2] // 60)
    seconds = int(list(item.values())[2] % 60)
    formatted_duration = f"{minutes}m{seconds}s"    #timestamp for every sentences

    sentence = item["text"]

    print(formatted_duration, "\t", sentence)


```
